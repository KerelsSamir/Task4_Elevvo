## **Exploratory Data Analysis (EDA) & Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(df.shape)
df.info()
df.describe()
```

### **Survival Analysis**

```python
sns.countplot(x='Survived', data=df, palette='rocket')
plt.title('Survival Counts (0 = Died, 1 = Survived)')
plt.show()
```

### **Survival by Sex**

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title('Survival by Sex')
plt.show()
```

### **Survival by Ticket Class (Pclass)**

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette='viridis')
plt.title('Survival by Passenger Class')
plt.show()
```

### **Survival Count by Embarkation Port**

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Embarked', hue='Survived', data=df, palette=['#FF9999','#66B2FF'])
plt.title('Survival Count by Embarkation Port (S = Southampton, C = Cherbourg, Q = Queenstown)')
plt.xlabel('Port of Embarkation')
plt.ylabel('Number of Passengers')
plt.show()
```

### **Fare Distribution by Survival**

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='Survived', y='Fare', data=df, palette='cool', inner='quartile')
plt.title('Fare Distribution by Survival')
plt.show()
```

### **Correlation Heatmap between Numerical Features**

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap between Numerical Features')
plt.show()
```
